```python
import numpy as np, pandas as pd
from sklearn.utils import shuffle
import matplotlib.pyplot as plt
import keras as K
from keras.models import Sequential
from keras.layers import Dropout, Dense
from keras.callbacks import EarlyStopping
from keras.optimizers import Adam
import matplotlib.pyplot as plt
import tensorflow


#read the data from the espicied file "data.xlsx"
file_name='data.xlsx'
data_excel = pd.DataFrame(pd.read_excel(file_name))
[n_rows,n_columns] =data_excel.shape
```

```python
#clean data
#print(data_excel['COLOR']) # all colors are 'Blanco' so they do not give any information to our model
data_excel= data_excel.drop(columns="COLOR")
data_excel['fecha']= data_excel['fecha'].apply(pd.to_datetime) #convert the date
data_excel['fecha'] = data_excel['fecha'] - data_excel['fecha'].min() #
#update columns
columns = data_excel.columns
# ETL data is normalized
for column in columns:
    if data_excel[column].dtypes == 'float64':
        data_excel[column] = (data_excel[column]-data_excel[column].min())/data_excel[column].max()
    else:
        data_excel[column] = data_excel[column].infer_objects()
        data_excel[column] = (data_excel[column]-data_excel[column].min())/data_excel[column].max()
data_excel=shuffle(data_excel)
```

```python
n_inputs=14
X=data_excel.iloc[:,0:n_inputs].as_matrix()
#Our goal is to catch the total mmbtu used so let's sum the last four columns.
Y=data_excel.iloc[:,19].as_matrix()
```

```python
model = Sequential()
model.add(Dense(100, input_shape=(n_inputs,)))
model.add(K.layers.LeakyReLU(alpha=0.03))
model.add(Dense(100))
model.add(Dropout(0.5))
model.add(K.layers.LeakyReLU(alpha=0.03))
model.add(Dense(1))

#model.add(K.layers.LeakyReLU(alpha=0.3))
#my_optimizer= K.optimizers.SGD(lr=0.0001, decay=1e-5, momentum=0.05, nesterov=True)
my_optimizer= K.optimizers.Adam(lr=0.0001)
model.compile(optimizer=my_optimizer, loss='logcosh', metrics=['logcosh'])
model.summary()
model.load_weights('weights_oven.h5')
```

```python
#print(xToPredict)
predichos=[]
reales=[]
#model.predict(xToPredict)
for number in range(0,n_rows,1):
    xToPredict= X[number].reshape((1,n_inputs))
    predichos.append(model.predict(xToPredict)[0][0])
    reales.append(Y[number])
#for number in range(0,57,1):
    #print("valores: {} , {}".format(reales[number], predichos[number]))
reales=pd.DataFrame(reales)
predichos = pd.DataFrame(predichos)
datos=pd.concat([reales,predichos],axis=1)
datos.to_csv(file_name+'_predictions.csv')
plt.plot(predichos,'g',label="predicho")
plt.plot(reales,'r',label="real")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
